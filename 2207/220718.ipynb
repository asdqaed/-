```python
import math

n = 123456*2 
a = [True]*(n+1)
a[0] = False
a[1] = False
s = []
for i in range(2, int(math.sqrt(n)) + 1): 
    if a[i] == True: 
        j = 2 
        while i * j <= n:
            a[i * j] = False
            j += 1

while True:
    x = int(input())
    if x == 0:
        break
    cnt = 0
    for i in range(x+1,(x*2)+1):
        if a[i]:
            cnt += 1
    s.append(cnt)
for i in s:
    print(i)
#정답1    
```

```python
import math

n = 10000
a = [True]*(n+1)
a[0] = False
a[1] = False
s = []
for i in range(2, int(math.sqrt(n)) + 1): 
    if a[i] == True:
        j = 2 
        while i * j <= n:
            a[i * j] = False
            j += 1
for i in range(n+1):
    if a[i]:
        s.append(i)

T = int(input())
for i in range(T):
    num = int(input())
    j1 = 0
    toggle = True
    min = [0,0]
    s_len = len(s)
    while s[j1]<num:
        j2 = 0
        while s[j2]<num:
            if s[j1]+s[j2] == num:
                if toggle:
                    min = [s[j1],s[j2]]
                    toggle = False
                elif abs(min[0]-min[1]) > abs(s[j1]-s[j2]):
                    min = [s[j1],s[j2]]
            j2 += 1
            if j2 == s_len:
                break
        j1 += 1
        if j1 == s_len:
            break
    print(min)
#시간초과
```
