```python
def jegui(k,n):
    if n==0:
        return 1
    elif n==k:
        return n
    else:
        return k*jegui(k+1,n)

print(jegui(1,int(input())))
```

```python
def jegui(k1,k2,cnt,n):
    if n == 1:
        return 1
    elif n == 0:
        return 0    
    elif cnt == n:
        return k2
    else:
        return jegui(k2,k1+k2,cnt+1,n)

print(jegui(1,1,2,int(input())))
```
