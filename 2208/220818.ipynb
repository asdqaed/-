```python
import sys
input = sys.stdin.readline

global white
global blue
white = 0
blue = 0

def cut_array(a,n):
    a_cut1 = []
    a_cut2 = []
    a_cut3 = []
    a_cut4 = []
    for i in range(n):
        a_cut1.append(a[i][:n])
        a_cut2.append(a[i][n:])
        a_cut3.append(a[n+i][:n])
        a_cut4.append(a[n+i][n:])
    result = [a_cut1,a_cut2,a_cut3,a_cut4]
    return result

def num_of_paper(a,n):
    global white
    global blue
    a_sum = 0
    for i in range(n):
        a_sum += sum(a[i])
    if a_sum == n*n or a_sum == 0:
        if a[0][0] == 1:
            blue += 1
        else:
            white += 1
    else:
        n_half = int(n/2)
        a_cut = cut_array(a,n_half)
        num_of_paper(a_cut[0],n_half)
        num_of_paper(a_cut[1],n_half)
        num_of_paper(a_cut[2],n_half)
        num_of_paper(a_cut[3],n_half)
n = int(input())
a = [[0]*n for i in range(n)]
for i in range(n):
    a[i] = list(map(int,input().split()))
num_of_paper(a,n)
print(white)
print(blue)
```
